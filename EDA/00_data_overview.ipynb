# 00. data overview

## Goal
Understand the datasets, keys, sizes, time coverage, and basic entities

## Why this matters
This step is important because before comparing business and consumer behavior, we need to understand what data we have, how the tables are linked, and what the main unit of analysis should be. 


```python
import pandas as pd
from mdq.data import load_data
```

```python
business, consumer, merchant, mcc = load_data()
```

```python
print('business:')
print(business.shape)

print()
print('consumer')
print(consumer.shape)

print()
print('merchant')
print(merchant.shape)
```

```python
print('business num of unique:')
print(business["card_number"].nunique())

print()
print('consumer num of unique:')
print(consumer["card_number"].nunique())

print()
print('merchant num of unique:')
print(merchant["merchant_id"].nunique())
```

```python
for name, df in [("business", business), ("consumer", consumer)]:
    ts = pd.to_datetime(df["transaction_timestamp"])
    print(name)
    print("min:", ts.min())
    print("max:", ts.max())
    print("days:", (ts.max() - ts.min()).days)
```

```python
overview = pd.DataFrame({
    "business": [
        len(business),
        business["card_number"].nunique(),
        business["merchant_id"].nunique(),
    ],
    "consumer": [
        len(consumer),
        consumer["card_number"].nunique(),
        consumer["merchant_id"].nunique(),
    ]
}, index=["transactions", "unique_cards", "unique_merchants"])

overview
```
